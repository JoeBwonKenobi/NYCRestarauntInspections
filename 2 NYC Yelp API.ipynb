```python
limit = 50
```

```python
results = yelp.search_query(term=term, location='Queens', limit=1)
total_results = results['total']
```

```python
limit = min(total_results, limit)
```

```python
num_batches = (total_results // limit) + 1
```

```python
max_results = 1000
```

```python
all_results = []
```

```python
# Calculate number of batches needed
num_batches = max_results // limit

for batch in range(num_batches):
    offset = batch * limit
    try:
        results = yelp.search_query(term=term, location=location, limit=limit, offset=offset, categories='pizza')
        all_results.extend(results['businesses'])
    except YelpAPI.YelpAPIError as e:
        print(f"An error occurred: {e}")
        break

# If there are any remaining results to fetch within the limit of 1000
remaining_results = max_results % limit
if remaining_results > 0:
    offset = num_batches * limit
    try:
        results = yelp.search_query(term=term, location=location, limit=remaining_results, offset=offset, categories='pizza')
        all_results.extend(results['businesses'])
    except YelpAPI.YelpAPIError as e:
        print(f"An error occurred: {e}")

print(f"Total results fetched: {len(all_results)}")
```

```python
# Save all results to the JSON file
with open(JSON_FILE, 'w') as f:
    json.dump(all_results, f)
```

```python
# Read the JSON file into a DataFrame
df = pd.read_json(JSON_FILE)
```
